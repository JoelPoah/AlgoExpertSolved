# Validate Subsequence
## Whether an array of numbers is in an array 

# solution 1

```python
from msilib import sequence
from random import randint
#generate an array of 1000 numbers between 1,100
array = [randint(1,100) for i in range(10000000)]
sequence = [randint(1,100) for i in range(1000)]
sequence
```

```python

def isValidSubsequence(array, sequence):
    total_yes =0 
    for i in range(len(array)):
        if array[i] == sequence[total_yes]:
            total_yes +=1
        if total_yes == len(sequence):
            return True
    return False
```

```python
isValidSubsequence(array, sequence)
```
